```python
import itertools
import random
import copy

# to change the number in row and to change the length of words, go to make_ok_combo
# to change from at most to exactly number in row, go to make_ok_combo

# list_of_words = ['FLAWED', 'CLOSED', 'ADMIRE', 'JUNIOR', 'BETRAY', 'PEANUT', 'MODEST', 'MAKEUP']
list_of_words = ['CHALK', 'POWER', 'BROWN', 'JUNIOR']


def main():
    combos, nums = make_all_word_combos(list_of_words)
    set_combos_for_participants(combos)

    
def make_all_word_combos(word_list):
    """
    This generates the okay combinations with letters for all 12 words, each using 
    a different seed number (the index in the word_list)
    """
    all_word_combos = []
    all_nums_in_a_row = list()
    for num in range(len(word_list)):
        print('NEW WORD')
        
        combos, nums_in_a_row = make_ok_combo(len(word_list[num]))
        
        all_word_combos.append(convert_combo_to_word(word_list[num], 
                make_ok_combo_random(seed_num=num, in_list=combos)))
        
        all_nums_in_a_row.append(nums_in_a_row)
        
    return all_word_combos, all_nums_in_a_row

def make_ok_combo(word_len):
    """
    This will generate the same list of ok combinations using numbers with at 
    most `number_in_row` number of numbers in a row.
    This is only generalizable to different word-lengths if 
    you change what l0 is set to (1-6 for 6 words)
    """
    
    # CHANGE THE NUMBER IN ROW HERE
    number_in_row = 1
    output = []
    nums_in_a_row = []
    
    l0 = list(itertools.permutations(list(range(1, word_len + 1))))

    # Converts l0 from a list of tuples to a list of lists, called l
    l = [list(permutation) for permutation in l0]
    

    # Makes a list of all the possible permutations with the first number added to the end 
    # (because they will be arranged in a circle), it's called l1
    l1 = [list(i) for i in l]
    for i in range(len(l1)):
        l1[i].append(l[i][0])

    # makes the output which is a list of ok combos with at most number_in_row number of 
    # numbers in a row
    for i in range(len(l1)):
        counter = 0
        for j in range(len(l1[i]) - 1):
            if (l1[i][j + 1] - l1[i][j] == 1) or (l1[i][j + 1] - l1[i][j] == -1):
                counter += 1
            # TO CHANGE FROM AT MOST NUMBER IN ROW TO EXACTLY NUMBER IN ROW, DO THE BELOW
            # if counter == number_in_row and j == (len(l1[i]) - 2): ## for if you want exactly 
            # that many in row
            if counter <= number_in_row and j == (len(l1[i]) - 2):
                nums_in_a_row.append(counter)
                output.append(l[i])
#                 the commented out line below would say how many are in a row
#                 output.append(counter)

    return output, nums_in_a_row


def make_ok_combo_random(seed_num, in_list):
    # this will take a seed number and a list, and will shuffle that list in a random order 
    # using the seed number
    # the in list will generally be the ok_combo_list
    # this is generalizable

    random.seed(seed_num)
    out_list = copy.deepcopy(in_list)
    random.shuffle(out_list)
    return out_list

def convert_combo_to_word(word, in_list):
    """
    this converts a list of lists of numbers (like an ok combo list) into having letters 
    instead of numbers.
    if the number is 1, that will be converted to the first letter of the word, etc.
    this is generalizable to different length words, as long as the length of the word 
    matches up with the presumed length of the word in the `in_list`.
    """

    out_list = copy.deepcopy(in_list)
    for i in range(0, len(in_list)):
        for j in range(len(in_list[i])):
            for num_cond in range(1, len(in_list[i]) + 1):
                if in_list[i][j] == num_cond:
                    out_list[i][j] = word[num_cond - 1]
    return out_list





def set_combos_for_participants(in_list):
    """
    This writes to a text file, for 240 participants each ok_combo (with letters) of 
    all the 12 words. No two combos are repeated, that's why there's only 240 participants. 
    And the combos are in random orders as specified by `make_ok_combo_random`.
    This is generalizable except that it assumes all the words have the same length 
    and number of characters in order.
    
    The number of outputs is as many as the minumum number of permutations provided by the 
    `in_list`, the code for this is found in the calculation for `total_number`. 
    """

    out_list = copy.deepcopy(in_list)
    
    total_number = len(min(out_list, key=len))
    
#     outfile = open('participant_words6.txt', 'w')
    outfile = open('participant_words5.txt', 'w')

    # this make an outlist with lists of strings instead of lists of lists, 
    # otherwise the same
    for participant_num in range(total_number):
        for word_num in range(len(out_list)):
            out_list[word_num][participant_num] = ''.join(out_list[word_num][participant_num])
    # uncomment below to see
    # return out_list

    # this writes to the file in such a way that it can easily be copy and pasted 
    # and work in the matlab code

    for participant_num in range(total_number):
        # TO START THE TXT FILE AT THE NTH PARTICIPANT, WRITE str(participant_num + 1 + N)
        outfile.write('Participant ' + str(participant_num + 1) + '\n')
        for word_num in range(len(out_list)):
            if word_num == len(out_list) - 1:
                outfile.write("'" + out_list[word_num][participant_num] + "'")
            else:
                outfile.write("'" + out_list[word_num][participant_num] + "',")
                # if we need a space between words, comment the above and uncomment the below
                # outfile.write("'" + out_list[word_num][participant_num] + "', ")
        outfile.write('\n')

    outfile.close()

main()
```
