```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
pd.set_option('display.max_columns', 50)

# Note that the total_data is out of order, it's participants 1 through 15, then participant 22, then participants
# 16 through 29 in the csv. 

total_data = pd.read_csv('total_participant_data.csv')
```

```python
# This is already transferred to the main notebook

def fix_participant_names(entry):
    if entry == 'P09':
        return 'P08'
    elif entry == 'P9':
        return 'P09'
    else:
        return entry

total_data.ParticipantName = total_data.ParticipantName.map(fix_participant_names)
```

```python
total_data.ParticipantName.unique()
```

```python
# Show a lineplot of the distance between the right eye and left eye X position for participant 1

sns.lineplot(data=total_data.loc[total_data.ParticipantName == 'P01']['GazePointRightX (ADCSpx)'] - \
             total_data.loc[total_data.ParticipantName == 'P01']['GazePointLeftX (ADCSpx)'])
```

```python
# make columns for the distance between the left and right eye X position and Y position 

total_data['gaze_X_diff'] = total_data['GazePointRightX (ADCSpx)'] - total_data['GazePointLeftX (ADCSpx)'] 
total_data['gaze_Y_diff'] = total_data['GazePointRightY (ADCSpx)'] - total_data['GazePointLeftY (ADCSpx)'] 
```

```python
# make a histogram of the distance between the left and right eye X position for each participant

total_data.groupby('ParticipantName')['gaze_X_diff'].hist(by=total_data.ParticipantName, range=(-200,200), bins=40)
```

```python
# make a histogram of the distance between the left and right eye Y position for each participant

total_data.groupby('ParticipantName')['gaze_Y_diff'].hist(by=total_data.ParticipantName, range=(-200,200), bins=40)
```

```python
avg_for_p01 = (total_data.loc[total_data.ParticipantName == 'P01']['GazePointLeftX (ADCSpx)'] + total_data.loc[total_data.ParticipantName == 'P01']['GazePointRightX (ADCSpx)']) // 2

avg_for_p01
```

```python
dif_between_gaze_avg = avg_for_p01 - total_data['GazePointX (ADCSpx)']

dif_between_gaze_avg.plot()
```
